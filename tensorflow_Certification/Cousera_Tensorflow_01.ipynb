신경망의 경우 작자는 두 숫자 사이의 관계를 학습하는 샘플을 예로 들었습니다. 예를 들어 이와 같은 함수에 대한 코드를 작성하는 경우 우리는 직관적으로 '규칙'을 알고 있습니다. 

float hw_function(float x){  
   float y = (2 * x) - 1;  
   return y;  
}  

그러면 이러한 작업을 수행하기 위해선 어떻게 신경망을 훈련시킬 수 있을까요?. 
그건 바로 데이터를 사용하는 겁니다. X 집합과 Y집합을 공급함으로써 그들 사이의 관계를 파악 하며 학습을 하계 됩니다.  

이것은 우리가 익숙하게 사용하던 방법과는 매우 다른 패러다임 이므로 한 단계씩 살펴보며 알아보도록 하겠습니다. 

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

### Import

import 부터 시작하겠습니다. 아래는 Tensorflow를 불러오고 그것을 쉽게 사용하기 위해 tf라고 하겠습니다.  

그런 다음 numpy라는 라이브러리를 가져와 데이터를 쉽고 빠르게 목록으로 표시 할 수 있게 됩니다.  

신경망을 일련의 순차 계층으로 정의하기 위한 프레임 워클르 keras라 부르고 이것또한 import 하도록 하겠습니다.  

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Define and Compile the Neural Network

다음 우리는 가장 간단한 신경망을 만들어 보도록 하겠습니다. 이것은 1개의 layer을 가지며, 또한 그 층은 1개의 뉴런을 가지며 그것의 input shape는 1개의 값을 갖습니다. 

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

다음으로 우리는 우리 신경망을 컴파일 하도록 하겠습니다 이것을 하는동안 우리는 2가지 기능을 지정하게 됩니다.    

만약 머신 러닝에 대해 알고 있다면 많은 수식들이, 여기서 사용되고 있다는 것을 알 수 있습니다. 하지만 이 tensorflow에선 이 함수들을 잘 요약하여 제공 하고 있습니다.  

우리는 이미 두 숫자에 대한 함수가 y = 2x-1 라는것을 알 고 있습니다,  

컴퓨터가 이것을 '학습'하기 위해선. 아마도 y=10x + 10. 과 같이 어떠한 값을 먼저 예측하게 됩니다. __LOSS function__ 은 알고있는 정답에 대한 추측된 답을 비교하고, 또 얼마나 잘 되고 있는지 나쁜지에 대해서 측정하게 됩니다.  

그 다음 __OPTIMIZER__ 함수를 사용하여 다른 추측을 하게됩니다.  
LOSS function이 어떻게 진행되었는지에 따라,  optimizer는 loss를 최소화 하려 합니다.  
그렇게 되면 y = 5x+5 와 같이 조금더 괜찮은 답을 내놓게 될것입니다. (즉 loss 값이 낮아집니다.)

이과정에 대해선 곧 보게될 EPOCH 수에 대해 이 과정을 반복하게 됩니다.  
그러나 여기선 먼저, MSQ( MEAN SQUARED ERROR) 최소 제곱법이 LOSS function에 사용되고 확률적 경사하강법 ( STOCHASTIC GRADIENT DESCENT )  기법이 optimizer에 사용 되었다는것만 알고 넘어가도록 하겠습니다.  

시간이 지남에 따라 우리는 다양하고 적절한 loss function과 optimizer function 에 대해 배우게 될 것입니다.  


```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

### Providing the Data

그다음 우리는 데이터를 주입 시켜보도록 하겠습니다. 여기선 6개의 x와 6개의 y값을 넣도록 하겠습니다. 우리는 이것들의 관계가 y=2x-1 임을 알고 있습니다.  

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Training the Nueral Network

X와 Y의 관계를 '학습'하는 신경망을 훈련시키는 과정은, model.fit 을 호출하면서 시작됩니다. 
여기서 우리가 언급한 루프를 거치며 값을 추축하고, 이것이 얼마나 좋고 나쁜지를 측정하게 되고(LOSS) 또 이것을 조정하기 위해 OPTIMIZER를 사용하게 됩니다. 이것을 우리가 지정한 EPOCH수 만큰 반복하게 되며, 아래 코드를 실행하면 현재 LOSS를 볼 수 있게 됩니다. 

```
model.fit(xs, ys, epochs=500)
```

```
print(model.predict([10.0]))
```
